### Regex - Find all the Numbers in a String

```python
"""
The Regex says 

r => raw string, i.e. the slashes are not for python but for the regex
slash d => 0-9 characters
+ => 1 or many
"""
senty = "there are 50 gold coins and in 7 bags"
import re
numbers = re.findall(r'\d+', senty)
numbers
```

### Lists 

```python
def age_sieve(age):
    if age < 18:
        return False
    else:
        return True
        
```

```python
ages = [5, 7, 17,  18,  24,  32]
print("ages =", ages)
adults = filter(age_sieve, ages)
print(adults)
print(list(adults))
```

## Tuple to List

The core difference between a Python list and a tuple is mutability: lists are mutable (changeable), while tuples are immutable (unchangeable). 

```python
some_tuple = (1, 2, 3, 4, 5)
print(type(some_tuple))
print(some_tuple)
```

```python
derived_list = list(some_tuple)
print(type(derived_list))
print(derived_list)
```

#### Updating a list and returning a new list

```python
base_list = ["anger", "frustration", "envy", "confusion"]
print("This will return None: ", base_list.append("healing"))
print("This will print the combibination: ", base_list +["peace"])
```

```python
def custom_comp(input_str: str) -> str:
    return len(input_str)
```

```python
# not in place update.. interesting that it removes duplicates.
sorted_list = sorted(base_list, key=custom_comp)
print(sorted_list)
```

### List Manipulation - filter, map and Lambdas

```python
def age_sieve(age: int) -> int:
    return age >= 18
```

```python
ages = [5, 7, 12, 17, 18, 24, 32]
adults = list(filter(age_sieve, ages))
print("adults := ", adults)
```

```python
def get_length(word: str) -> int:
    return len(word)
```

```python
fruit_lenghths = list(map(get_length, ["apple", "banana", "cherry", "durien"]))
print("fruit lengths: ", fruit_lenghths)
```
